```python
import pandas as pd
ids = pd.read_csv('imdb_ids.csv', header=None)
```

```python
id_list = []
for i in ids[0]:
    x = "{:>07}".format(str(i))
    id_list.append(x)
```

```python
id_list
```

```python
#feldarabolás, 5 000 lesz egy listában
chunks = [id_list[x:x+5000] for x in range(0, len(id_list), 5000)]
```

```python
chunks
```

```python
df_list = []

for i in chunks:
    df = pd.DataFrame(i)
    df_list.append(df)
```

```python
df_list
```

```python
#kiírás csv-be, i-t használja, hogy elnevezze a fájlokat (különbözőek legyenek)
import os

csv = "csv"
for i, df in enumerate(df_list, 1):
    filename = "imdb_ids_{}.{}".format(i, csv)
    df.to_csv(filename, index=False, header = False)
```
